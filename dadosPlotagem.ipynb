```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Nome':['Alice','Joao','Joana','Ana','Gustavo'],
        'Idade':[25, 30, 45, 23, 50],
        'Salario':[5000, 2000, 7000, 11000, 8000]}
```

```python
df = pd.DataFrame(data)
```

```python
#Plotar um codigo de barras para salarios

plt.bar(df['Nome'], df['Salario'], color='blue')
plt.title('Gráfico de Barras - Salários')

plt.xlabel('Nome')
plt.ylabel('Salário (R$)')
plt.show()
```

```python
# Grafico de Pizza (autopct eh quantos zeros dps da virgula)
plt.pie(df['Salario'], labels=df['Nome'], autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])
plt.title('Grafico Pizza Salários')

plt.show()
```

```python
# Grafico de Pizza (autopct eh quantos zeros dps da virgula)
plt.pie(df['Salario'], labels=df['Nome'], autopct='%1.1f%%')
plt.title('Grafico Pizza Salários')

plt.show()
```

```python
# Grafico de Dispersao
plt.scatter(df['Idade'], df['Salario'], color='purple')
plt.title('Gráfico de Dispersão Idade X Salário')

plt.xlabel('Idade')
plt.ylabel('Salário R$')

plt.show()
```

```python
# Grafico de Linhas
plt.plot(df['Nome'], df['Idade'], marker='o', linestyle='-', color='green')
plt.title('Grafico de Linhas - Idades')
plt.xlabel('Nome')
plt.ylabel('Idade')

plt.show()
```

```python
# Grafico de Linhas para Salario
plt.plot(df['Nome'], df['Salario'], marker='*', linestyle='-', color='pink')
plt.title('Grafico de Linhas - Salário')
plt.xlabel('Nome')
plt.ylabel('Salario')

plt.show()
```
